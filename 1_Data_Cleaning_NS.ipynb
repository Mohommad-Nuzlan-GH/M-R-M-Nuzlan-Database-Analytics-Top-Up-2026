```python
# Import pandas for data handling
import pandas as pd

# Load all CSV files using entity names
CUSTOMER = pd.read_csv('1#customers.csv')
APP_EVENT = pd.read_csv('2#app_events.csv')
ORDER = pd.read_csv('3#orders.csv')
COMPLAINT = pd.read_csv('4#complaints.csv')
DELIVERY = pd.read_csv('5#deliveries.csv')
DRIVER = pd.read_csv('6#drivers.csv')
VEHICLE = pd.read_csv('7#vehicles.csv')
HUB = pd.read_csv('8#hubs.csv')
INCIDENT = pd.read_csv('9#incidents.csv')
```

```python
#--------------------CUSTOMER CLEANING------------------------------
```

```python
#1 Check CUSTOMER structure
CUSTOMER.info()
CUSTOMER.head()
```

```python
#2 Check missing values in CUSTOMER
CUSTOMER.isnull().sum()
```

```python
#3 Remove duplicate rows
CUSTOMER = CUSTOMER.drop_duplicates()
```

```python
#4 Standardise text columns
CUSTOMER['home_zone'] = CUSTOMER['home_zone'].str.strip().str.title()
CUSTOMER['home_zone'] = CUSTOMER['home_zone'].replace({'Ctr': 'Central'})

CUSTOMER['customer_type'] = CUSTOMER['customer_type'].str.strip().str.title()
CUSTOMER['preferred_channel'] = CUSTOMER['preferred_channel'].str.strip()
CUSTOMER['account_status'] = CUSTOMER['account_status'].str.strip().str.title()
```

```python
#5 Convert signup date into datetime format
CUSTOMER['signup_date'] = pd.to_datetime(CUSTOMER['signup_date'], errors='coerce')
```

```python
#6.1 Fill missing loyalty score using median
CUSTOMER['loyalty_score'] = CUSTOMER['loyalty_score'].fillna(CUSTOMER['loyalty_score'].median())

#6.2 Fill missing preferred channel using Unknown
CUSTOMER['preferred_channel'] = CUSTOMER['preferred_channel'].fillna('Unknown')
```

```python
#7 Create customer age group column
CUSTOMER['age_group'] = pd.cut(
    CUSTOMER['age'],
    bins=[0, 25, 40, 60, 100],
    labels=['Young', 'Adult', 'Middle_Aged', 'Senior']
)
```

```python
#--------------------APP_EVENT CLEANING------------------------------
```

```python
#1 Check APP_EVENT structure
APP_EVENT.info()
APP_EVENT.head()
```

```python
#2 Check missing values in APP_EVENT
APP_EVENT.isnull().sum()
```

```python
#3 Remove duplicate rows
APP_EVENT = APP_EVENT.drop_duplicates()
```

```python
#4 Standardise text columns
APP_EVENT['event_type'] = APP_EVENT['event_type'].str.strip().str.lower()
APP_EVENT['device_type'] = APP_EVENT['device_type'].str.strip()

APP_EVENT['zone_context'] = APP_EVENT['zone_context'].str.strip().str.title()
APP_EVENT['zone_context'] = APP_EVENT['zone_context'].replace({'Ctr': 'Central'})
```

```python
#5 Convert event timestamp into datetime format
APP_EVENT['event_timestamp'] = pd.to_datetime(APP_EVENT['event_timestamp'], errors='coerce')
```

```python
#6 Handle missing order_id
APP_EVENT['order_id'] = APP_EVENT['order_id'].fillna('No_Order_Link')
```

```python
#7 Create event success status column
APP_EVENT['event_success_status'] = APP_EVENT['success_flag'].map({
    1: 'Success',
    0: 'Failed'
})
```

```python
#8 Create latency category column
APP_EVENT['latency_category'] = pd.cut(
    APP_EVENT['api_latency_ms'],
    bins=[0, 100, 300, 10000],
    labels=['Low', 'Medium', 'High']
)
```

```python
#------------------------ORDER CLEANING-----------------------------------
```

```python
#1 Check ORDER structure
ORDER.info()
ORDER.head()
```

```python
#2 Check missing values in ORDER
ORDER.isnull().sum()
```

```python
#3 Remove duplicate rows
ORDER = ORDER.drop_duplicates()
```

```python
#4 Standardise text columns
ORDER['service_type'] = ORDER['service_type'].str.strip().str.title()
ORDER['pickup_zone'] = ORDER['pickup_zone'].str.strip().str.title()
ORDER['dropoff_zone'] = ORDER['dropoff_zone'].str.strip().str.title()

ORDER['pickup_zone'] = ORDER['pickup_zone'].replace({'Ctr': 'Central'})
ORDER['dropoff_zone'] = ORDER['dropoff_zone'].replace({'Ctr': 'Central'})

ORDER['priority_level'] = ORDER['priority_level'].str.strip().str.title()
ORDER['booking_channel'] = ORDER['booking_channel'].str.strip()
```

```python
#5 Convert order created date into datetime format
ORDER['order_created_at'] = pd.to_datetime(ORDER['order_created_at'], errors='coerce')
```

```python
#6 Fill missing booking channel using Unknown
ORDER['booking_channel'] = ORDER['booking_channel'].fillna('Unknown')
```

```python
#7 Create high value order flag
ORDER['high_value_order'] = ORDER['order_value'] > ORDER['order_value'].median()
```

```python
#-----------------COMPLAINT CLEANING--------------------------------
```

```python
#1 Check COMPLAINT structure
COMPLAINT.info()
COMPLAINT.head()
```

```python
#2 Check missing values in COMPLAINT
COMPLAINT.isnull().sum()
```

```python
#3 Remove duplicate rows
COMPLAINT = COMPLAINT.drop_duplicates()
```

```python
#4 Standardise text columns
COMPLAINT['complaint_type'] = COMPLAINT['complaint_type'].str.strip().str.title()
COMPLAINT['channel'] = COMPLAINT['channel'].str.strip()
COMPLAINT['severity'] = COMPLAINT['severity'].str.strip().str.title()
COMPLAINT['status'] = COMPLAINT['status'].str.strip().str.title()
```

```python
#5 Convert complaint created date into datetime format
COMPLAINT['created_at'] = pd.to_datetime(COMPLAINT['created_at'], errors='coerce')
```

```python
#6 Fill missing compensation amount with 0
COMPLAINT['compensation_amount'] = COMPLAINT['compensation_amount'].fillna(0)
```

```python
#7 Create complaint resolved flag
COMPLAINT['is_resolved'] = COMPLAINT['status'].apply(
    lambda x: 1 if x == 'Resolved' else 0
)
```

```python
#-------------------DELIVERY CLEANING----------------------------------
```

```python
#1 Check DELIVERY structure
DELIVERY.info()
DELIVERY.head()
```

```python
#2 Check missing values in DELIVERY
DELIVERY.isnull().sum()
```

```python
#3 Remove duplicate rows
DELIVERY = DELIVERY.drop_duplicates()
```

```python
#4 Standardise text columns
DELIVERY['delivery_status'] = DELIVERY['delivery_status'].str.strip().str.title()
```

```python
#5 Convert delivery dates into datetime format
DELIVERY['dispatch_time'] = pd.to_datetime(DELIVERY['dispatch_time'], errors='coerce')
DELIVERY['delivery_completed_at'] = pd.to_datetime(DELIVERY['delivery_completed_at'], errors='coerce')
```

```python
#6 Fill missing customer ratings using median
DELIVERY['customer_rating_post_delivery'] = DELIVERY['customer_rating_post_delivery'].fillna(
    DELIVERY['customer_rating_post_delivery'].median()
)
```

```python
#7 Keep missing completion time as blank (for failed deliveries)
# No fill applied to delivery_completed_at
```

```python
#8 Create delivery duration in hours
DELIVERY['delivery_duration_hours'] = (
    DELIVERY['delivery_completed_at'] - DELIVERY['dispatch_time']
).dt.total_seconds() / 3600
```

```python
#9 Create failed delivery flag
DELIVERY['failed_delivery_flag'] = DELIVERY['delivery_status'].apply(
    lambda x: 1 if x == 'Failed' else 0
)
```

```python
#10 Create delayed delivery flag
DELIVERY['delayed_delivery_flag'] = DELIVERY['delivery_status'].apply(
    lambda x: 1 if x == 'Delayed' else 0
)
```

```python
#-------------------DRIVER CLEANING------------------------------
```

```python
#1 Check DRIVER structure
DRIVER.info()
DRIVER.head()
```

```python
#2 Check missing values in DRIVER
DRIVER.isnull().sum()
```

```python
#3 Remove duplicate rows
DRIVER = DRIVER.drop_duplicates()
```

```python
#4 Standardise text columns
DRIVER['base_zone'] = DRIVER['base_zone'].str.strip().str.title()
DRIVER['base_zone'] = DRIVER['base_zone'].replace({'Ctr': 'Central'})

DRIVER['employment_type'] = DRIVER['employment_type'].str.strip().str.title()
DRIVER['shift_preference'] = DRIVER['shift_preference'].str.strip().str.title()
```

```python
#5 Fill missing training score using median
DRIVER['training_score'] = DRIVER['training_score'].fillna(
    DRIVER['training_score'].median()
)
```

```python
#6 Create driver experience level column
DRIVER['experience_level'] = pd.cut(
    DRIVER['years_experience'],
    bins=[-1, 2, 5, 10, 50],
    labels=['New', 'Intermediate', 'Experienced', 'Senior']
)
```

```python
#-----------------------------VEHICLE CLEANING-----------------------
```

```python
#1 Check VEHICLE structure
VEHICLE.info()
VEHICLE.head()
```

```python
#2 Check missing values in VEHICLE
VEHICLE.isnull().sum()
```

```python
#3 Remove duplicate rows
VEHICLE = VEHICLE.drop_duplicates()
```

```python
#4 Standardise text columns
VEHICLE['vehicle_type'] = VEHICLE['vehicle_type'].str.strip().str.title()

VEHICLE['assigned_zone'] = VEHICLE['assigned_zone'].str.strip().str.title()
VEHICLE['assigned_zone'] = VEHICLE['assigned_zone'].replace({'Ctr': 'Central'})

VEHICLE['maintenance_status'] = VEHICLE['maintenance_status'].str.strip().str.title()
VEHICLE['telematics_version'] = VEHICLE['telematics_version'].str.strip()
```

```python
#5 Convert commission date into datetime format
VEHICLE['commission_date'] = pd.to_datetime(VEHICLE['commission_date'], errors='coerce')
```

```python
#6 Fill missing battery health using median
VEHICLE['battery_health_pct'] = VEHICLE['battery_health_pct'].fillna(
    VEHICLE['battery_health_pct'].median()
)
```

```python
#7 Create battery health category
VEHICLE['battery_health_category'] = pd.cut(
    VEHICLE['battery_health_pct'],
    bins=[0, 60, 80, 100],
    labels=['Low', 'Medium', 'High']
)
```

```python
#----------------------------HUB CLEANING-------------------------------
```

```python
#1 Check HUB structure
HUB.info()
HUB.head()
```

```python
#2 Check missing values in HUB
HUB.isnull().sum()
```

```python
#3 Remove duplicate rows
HUB = HUB.drop_duplicates()
```

```python
#4 Standardise text columns
HUB['hub_name'] = HUB['hub_name'].str.strip().str.title()
HUB['zone'] = HUB['zone'].str.strip().str.title()
HUB['hub_type'] = HUB['hub_type'].str.strip().str.title()
```

```python
#5 Create hub capacity category
HUB['capacity_category'] = pd.cut(
    HUB['capacity_score'],
    bins=[0, 50, 75, 100],
    labels=['Low', 'Medium', 'High']
)
```

```python
#--------------------INCIDENT CLEANING--------------------------------
```

```python
#1 Check INCIDENT structure
INCIDENT.info()
INCIDENT.head()
```

```python
#2 Check missing values in INCIDENT
INCIDENT.isnull().sum()
```

```python
#3 Remove duplicate rows
INCIDENT = INCIDENT.drop_duplicates()
```

```python
#4 Standardise text columns
INCIDENT['incident_type'] = INCIDENT['incident_type'].str.strip().str.title()
INCIDENT['severity'] = INCIDENT['severity'].str.strip().str.title()
INCIDENT['resolution_status'] = INCIDENT['resolution_status'].str.strip().str.title()
```

```python
#5 Convert reported date into datetime format
INCIDENT['reported_at'] = pd.to_datetime(INCIDENT['reported_at'], errors='coerce')
```

```python
#6 Fill missing resolved hours with median
INCIDENT['resolved_hours'] = INCIDENT['resolved_hours'].fillna(
    INCIDENT['resolved_hours'].median()
)
```

```python
#7 Create unresolved incident flag
INCIDENT['unresolved_flag'] = INCIDENT['resolution_status'].apply(
    lambda x: 1 if x in ['Open', 'Escalated', 'Pendingvendor'] else 0
)
```

```python
# Final check for missing values and duplicates in all datasets

datasets = {
    "CUSTOMER": CUSTOMER,
    "APP_EVENT": APP_EVENT,
    "ORDER": ORDER,
    "COMPLAINT": COMPLAINT,
    "DELIVERY": DELIVERY,
    "DRIVER": DRIVER,
    "VEHICLE": VEHICLE,
    "HUB": HUB,
    "INCIDENT": INCIDENT
}

for name, df in datasets.items():
    print(f"\n{name} DATASET CHECK")

    # Check missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Check duplicates
    print("\nDuplicate Rows:", df.duplicated().sum())
```
